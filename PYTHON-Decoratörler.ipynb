```python
import logging
from functools import wraps
import time
logging.basicConfig(
    filename="islem_loglari.txt",  # log dosyası adı
    level=logging.INFO,            # INFO ve üstü seviyeleri kaydet
    format="%(asctime)s - %(levelname)s - %(message)s"
)
def log_islem(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
         try:
            logging.info(f"{func.__name__} fonksiyonu çağrıldı.")
            start = time.perf_counter()
            result = func(*args, **kwargs)  # fonksiyon çalıştırılıyor
            end = time.perf_counter()
            sure = end - start
            print(f"{sure} sürede çalışmayı tamamladı.")
            logging.info(f"{func.__name__} fonksiyonu başarıyla tamamlandı.")
            return result
         except Exception as e:
            logging.error(f"{func.__name__} fonksiyonunda hata aldı")
            print("Hata: ", e)
    return wrapper
@log_islem
def printer():
    yas = int(input("yaş giriniz"))
    print("merhaba", yas)
```

```python
printer()
```

```python
from functools import wraps

import time
def a(f):
    @wraps(f)
    def wrper(*args, **kwargs):
        start = time.perf_counter()
        result = f(*args, **kwargs)
        end = time.perf_counter()
        elapsed = end - start
        print(f"{f.__name__} fonksiyonu {elapsed:.3f} saniye çalıştı")
        return result  
    return wrper
@a
def merhaba():
    sayi = 0
    for x in range(1000000):
        sayi +=x
    else:
        return sayi
print(merhaba())
```
