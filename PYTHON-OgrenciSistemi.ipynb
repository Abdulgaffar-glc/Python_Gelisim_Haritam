```python
ogrenci1 = {}
liste = []
def ogr_ekle():
    ogrenci = ogrenci1.copy()
    try:   
        ogrenci["id"] = len(liste)
        ogrenci["ad"] = input("Adını giriniz")
        ogrenci["soyad"] = input("Soyadını giriniz")
        ogrenci["vize_notu"] = int(input("Vize notunu giriniz"))
        ogrenci["final_notu"] = int(input("Final notunu giriniz"))
    except:
        print("Geçersiz değer girdiniz...")
    else:
        liste.append(ogrenci)
ogr_ekle()
```

```python
ogr_ekle()
```

```python
def ogr_ara():
    aranan_ad = input("Aramak istediğiniz öğrencinin adını giriniz")
    aranan_soyad = input("Aramak istediğiniz öğrencinin soyadını giriniz")
    sonuc = [x if x!= None else print("öğrenci bulunamadı") for x in liste if x["ad"]== aranan_ad and x["soyad"] == aranan_soyad]
    if sonuc:
        print(sonuc)
    else:
        print("Ogrenci bulunamadi..")
```

```python
ogr_ara()
```

```python
def ogr_listele():
    for ogrr in liste:
        for ogr_key in ogrr:
            print(f"{ogr_key} , {ogrr[ogr_key]}")
        else:
            print()
```

```python
ogr_listele()
```

```python
def ogr_sil():
    ogr_listele()
    try:
        sil_no = int(input("Silmek istediğiniz öğrenci id'sini giriniz.."))
    except ValueError:
        print("Hatalı bir değer girdiniz")
    for x in liste:
        if x["id"] == sil_no:
            liste.remove(x)
            print("Öğrenci silindi..")
            break
    else:
        print("Ögrenci bulunamadı..")
```

```python
def harf_not_getir(ort_not):
    if ort_not>=90:
        return "AA"
    elif ort_not>=80:
        return "BA"
    elif ort_not>=70:
        return "BB"
    elif ort_not>=60:
        return "CB"
    elif ort_not>=50:
        return "CC"
    elif ort_not>=40:
        return "DC"
    elif ort_not>=40:
        return "DD"
    else:
        return "FF"
def harf_notu():
    for ogr in liste:
        ortalama = (ogr["vize_notu"]+ogr["final_notu"])/2
        harf_not_getir(ortalama)
        ogr["Ortalama"] = int(ortalama)
        ogr["Harf_notu"] = harf_not_getir(ortalama)
        for ogr_key in ogr:
            print(f"{ogr_key} , {ogr[ogr_key]}")
        else:
            print()
```
