```python
import logging
from functools import wraps
import time
logging.basicConfig(
    filename="islem_loglari.txt",  # log dosyası adı
    level=logging.INFO,            # INFO ve üstü seviyeleri kaydet
    format="%(asctime)s - %(levelname)s - %(message)s"
)
def log_islem(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
         try:
            logging.info(f"{func.__name__} fonksiyonu çağrıldı.")
            start = time.perf_counter()
            result = func(*args, **kwargs)  # fonksiyon çalıştırılıyor
            end = time.perf_counter()
            sure = end - start
            print(f"{sure} sürede çalışmayı tamamladı.")
            logging.info(f"{func.__name__} fonksiyonu başarıyla tamamlandı.")
            return result
         except:
            logging.error(f"{func.__name__} fonksiyonunda hata aldı")
            raise  # hatayı tekrar fırlat ki programda belli olsun
    return wrapper
```

```python
import csv

def birlestir(dosya1, dosya2, cikti):
    with open(dosya1, newline='') as f1, \
         open(dosya2, newline='') as f2:
        
        veri1 = list(csv.DictReader(f1))
        veri2 = list(csv.DictReader(f2))
        
        # sütun adlarını kontrol et
        sutun1 = veri1[0].keys()
        sutun2 = veri2[0].keys()
    
    with open(cikti, "w", newline='') as f_out:
        if sutun1 == sutun2:  
            # Aynı yapıdaysa alt alta ekle
            yazici = csv.DictWriter(f_out, fieldnames=sutun1)
            yazici.writeheader()
            yazici.writerows(veri1 + veri2)
        else:
            # Farklıysa yan yana ekle
            max_len = max(len(veri1), len(veri2))
            # Eksik satırları {} ile dolduralım
            veri1.extend([{}]*(max_len-len(veri1)))
            veri2.extend([{}]*(max_len-len(veri2)))
            
            # Sütunları birleştir
            fieldnames = list(sutun1) + list(sutun2)
            yazici = csv.DictWriter(f_out, fieldnames=fieldnames)
            yazici.writeheader()
            for row1, row2 in zip(veri1, veri2):
                yeni = {**row1, **row2}
                yazici.writerow(yeni)

    print("Dosyalar birleştirildi ->", cikti)
```

```python
class Drone:
    def __init__(self , id , model , max_speed , speed , battary, location):
        self.id = id
        self.model= model
        self.max_speed = max_speed
        self.speed = speed
        self.__battary = battary
        self.location= location
    def set_battary(self, new_bat):
        if new_bat <=10:
            print("şarj bitiyor güç takviye edin...")
        else:
            self.__battary = new_bat
    def status(self):
        self.battary = self.get_battary()
        print(f"{self.id} modeliniz {self.model}  max hızınız {self.max_speed} hızınız {self.speed}  bataryanız {self.get_battary} konumunuz {self.location}")
    def get_battary(self):
        return self.__battary
    def fly(self, yeni_konum):
        self.set_battary(50)
        self.speed +=50
        print(f"{self.location} konumundan, {yeni_konum}'a hareket edildi")
        self.location = yeni_konum
        if self.battary<10:
            print("Geri dönün güç kritik durumda...")
akinci= Drone(1,max_speed=300,model= "İHA", speed=250, battary=100, location=588.65)
kizielma = Drone(2,max_speed=1000,model="SİHA", speed=790, battary=100, location=448.65)
```

```python
from abc import ABC, abstractmethod
class shape(ABC):
    def __init__(self,kenar):
         self.kenar = kenar
    @abstractmethod
    def hesapla():
        pass
class Daire(shape):
    def __init__(self,kenar):
        super().__init__(kenar=kenar)
    def hesapla(self):
        sonuc = 3.14* self.kenar **2
        print("alan = ",sonuc)
class Kare(shape):
    def __init__(self,kenar):
        super().__init__(kenar)
    def hesapla(self):
        sonuc = self.kenar **2
        print("alan = ",sonuc)
class Dikdortgen(shape):
    def __init__(self,kenar1, kenar2):
        super().__init__(kenar1)
        self.kenar2 = kenar2
    def hesapla(self):
        sonuc = self.kenar*self.kenar2
        print("alan = ",sonuc)
```

```python
dar = Daire(20)
dar.hesapla()
dik = Dikdortgen(20,15)
dik.hesapla()
kar = Kare(20)
kar.hesapla()
```

```python
import numpy as np
from datetime import datetime
class banka:
    def __init__(self,isim,soyisim,bakiye, hesap_no=0):
        self.isim = isim
        self.soyisim = soyisim
        self.bakiye = bakiye
        self.hesap_no = np.random.randint(10000000,1000000000)
    @property
    def bakiye(self):
        return self.bakiye
    @bakiye.setter
    def bakiye(self,bakiye):
        if bakiye<0:
            raise ValueError("Geçersiz bir değer girildi")
        else:
            self._bakiye = bakiye
    def __str__(self):
        return f"{self.hesap_no} nolu {self.isim} {self.soyisim} isimli kişinin hesabıın bakiyesi {self._bakiye} Tl "
    def para_yatir(self, miktar):
        self._bakiye += miktar
        print(f"Güncel bakiyeniz {self._bakiye}")
        print("Başarılı")
        with open("işlem.txt","a", encoding="Utf-8")as data:
            data.write(f"{datetime.now()} başarılı işlem hesaba {miktar} tl yatırıldı Güncel bakiye {self._bakiye}\n")
    def para_cek(self, miktar):
        if self._bakiye < miktar:
            print("yetrsiz bakiye")
            with open("işlem.txt","a", encoding="Utf-8") as data:
                data.write(f"{datetime.now()} başarısız işlem hesaba {miktar} tl yatırılamadı, bakiye {self._bakiye}\n")
                print(f"Mevcut bakiyeniz {self._bakiye}")
        else:
            self._bakiye -= miktar
            print(f"Güncel bakiyeniz {self._bakiye}")
            print("başarılı")
            with open("işlem.txt","a", encoding="Utf-8") as data:
                data.write(f"{datetime.now()} başarılı işlem hesaba {miktar} tl çekildi Güncel bakiye {self._bakiye}\n")
```

```python
ahme = banka(isim="ahmet",soyisim="gülice", bakiye= 50)
```

```python
print(ahme)
```

```python
ahme.para_yatir(500)
```

```python
ahme.para_cek(550)
```
