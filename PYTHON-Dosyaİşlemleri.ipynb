```python
import csv
def filtrele(dosya_adi,sinir):
    with open(dosya_adi, "r") as dosya:
        verii = csv.reader(dosya)
        for no,a in enumerate(verii):
            if no == 0:
                continue
            if sinir < float(a[3]):
                    print(a)
```

```python
import csv

def birlestir(dosya1, dosya2, cikti):
    with open(dosya1, newline='') as f1, \
         open(dosya2, newline='') as f2:
        
        veri1 = list(csv.DictReader(f1))
        veri2 = list(csv.DictReader(f2))
        
        # sütun adlarını kontrol et
        sutun1 = veri1[0].keys()
        sutun2 = veri2[0].keys()
    
    with open(cikti, "w", newline='') as f_out:
        if sutun1 == sutun2:  
            # Aynı yapıdaysa alt alta ekle
            yazici = csv.DictWriter(f_out, fieldnames=sutun1)
            yazici.writeheader()
            yazici.writerows(veri1 + veri2)
        else:
            # Farklıysa yan yana ekle
            max_len = max(len(veri1), len(veri2))
            # Eksik satırları {} ile dolduralım
            veri1.extend([{}]*(max_len-len(veri1)))
            veri2.extend([{}]*(max_len-len(veri2)))
            
            # Sütunları birleştir
            fieldnames = list(sutun1) + list(sutun2)
            yazici = csv.DictWriter(f_out, fieldnames=fieldnames)
            yazici.writeheader()
            for row1, row2 in zip(veri1, veri2):
                yeni = {**row1, **row2}
                yazici.writerow(yeni)

    print("Dosyalar birleştirildi ->", cikti)
```

```python
birlestir("kitaplar.csv","kitaplar1.csv","cikti.csv")
```
